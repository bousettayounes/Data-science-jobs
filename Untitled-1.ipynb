# (exploring each columns)

experience level

```python
Data['experience_level'] = Data['experience_level'].replace('EN','Entry-level/Junior')
Data['experience_level'] = Data['experience_level'].replace('MI','Mid-level/Intermediate')
Data['experience_level'] = Data['experience_level'].replace('SE','Senior-level/Expert')
Data['experience_level'] = Data['experience_level'].replace('EX','Executive-level/Director')
```

```python
level = Data['experience_level'].value_counts()
level
```

job title

```python
print(' job titles in the dataset is : ',Data['job_title'].value_counts().size)
```

```python
top10_job_title = Data['job_title'].value_counts()[:10]
fig = px.bar(y=top10_job_title.values, 
             x=top10_job_title.index, 
             color = top10_job_title.index,
             #color_discrete_sequence=px.colors.sequential.PuBuGn,
             text=top10_job_title.values,
             title= 'Top 10 Job Titles',
             template= 'presentation')
fig.update_layout(
    xaxis_title="Job Titles",
    yaxis_title="count",
    font = dict(size=17,family="Franklin Gothic"))
fig.show()
```

Employment Type



```python
Data['employment_type'].value_counts()
```

```python
Data['employment_type'] = Data['employment_type'].replace('FT','Full Time')
Data['employment_type'] = Data['employment_type'].replace('PT','Part-Time')
Data['employment_type'] = Data['employment_type'].replace('CT','Contract')
Data['employment_type'] = Data['employment_type'].replace('FL','Freelance')
```

```python
type_grouped = Data['employment_type'].value_counts()
emp_type = ['Full-Time','Part-Time','Contract','Freelance']
fig = px.bar(x = emp_type, y = type_grouped.values, 
       color = type_grouped.index, 
       template = 'presentation',
       text = type_grouped.values, title = 'Employment Type Distribution')
fig.update_layout(
    xaxis_title="Employment Type",
    yaxis_title="count",
    font = dict(size=17,family="Franklin Gothic"))
fig.update_traces(width=0.5)
fig.show()
```

# Salary Analysis

```python
w2020 = Data.loc[(Data['work_year'] == 2020)]
w2021 = Data.loc[(Data['work_year'] == 2021)]
w2022 = Data.loc[(Data['work_year'] == 2022)]


year_salary = pd.DataFrame(columns=['2020','2021','2022'])
year_salary['2020'] = w2020.groupby('work_year').mean('salary_in_usd')['salary_in_usd'].values
year_salary['2021'] = w2021.groupby('work_year').mean('salary_in_usd')['salary_in_usd'].values
year_salary['2022'] = w2022.groupby('work_year').mean('salary_in_usd')['salary_in_usd'].values
```

```python
year_salary
```
